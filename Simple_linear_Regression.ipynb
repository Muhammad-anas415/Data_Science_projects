```python
#Importing the libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import root_mean_squared_error
import pickle
```

```python
#loading the dataset
df=pd.read_csv("D:\DATA_SCIENCE_PROJECTS\DATASETS\Salary_Data.csv")
df.head()
```

```python
#Drawing a scatter plot using seaborn library
sns.scatterplot(x="YearsExperience", y="Salary", data=df)
```

Since the dataset shows a clear linear relation between experience and salary, we apply simple linear regression.



#### Creating X and Y


```python
#converting the data into x and y
x=df[["YearsExperience"]]
y=df["Salary"]
```

#### Splitting the dataset into train and test sets

```python
#splitting the data into train  
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

#### Training the model

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

### checking the prediction line in the scatterplot 

```python
#creating a scatterplot for the prediction line 
plt.scatter(x_train,y_train,color="green")
plt.plot(x_train,model.predict(x_train),color="red")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

### checking the accuracy score of the model 

```python
#cheking the score of the model 
score=model.score(x_test,y_test)
round_score = round(score, 2)
print(f"The score of the model is {round_score}")
```

### checking the score of the model using the metrics

```python
#chekcing the score of model using the metrics
y_pred = model.predict(x_test)

r2 = r2_score(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

# rounding
r2 = round(r2, 3)
rmse = round(rmse, 3)
mse = round(mse, 3)

print(f"The RÂ² score of the model is: {r2}")
print(f"The RMSE of the model is: {rmse}")
print(f"The MSE of the model is: {mse}")
```

### Making in user interface for the prediction of salary based on experience

```python
User_input = float(input("Enter the years of experience: "))
prediction = model.predict([[User_input]])
prediction_rounded = round(prediction[0], 2)   # round to 2 decimal places
print(f"The predicted salary for {User_input} years of experience is: {prediction_rounded}")
   
```
